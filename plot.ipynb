```python
import pandas as pd
from sys import argv
from matplotlib import pyplot as plt
```

```python
for test in ["test1", "test2", "test3"]:
    for lr, ef in [
        (0.1, 0.1), (0.1, 0.5), (0.1, 0.9), 
        (0.5, 0.1), (0.5, 0.5), (0.5, 0.9), 
        (0.9, 0.1), (0.9, 0.5), (0.9, 0.9)]:  
        filename = test + "_" + str(lr) + "_" + str(ef)
        
        file = pd.read_csv(filename)
        x = file.iloc[:, 0].to_list()
        y = file.iloc[:, 1].to_list()

        plt.plot(x, y, label="lr="+str(lr)+",ef="+str(ef))
        
    plt.title("Média de recompensas vs Episódios (" + test + ")")                
    plt.xlabel("Episódios")
    plt.ylabel("Média de recompensas")
    plt.legend(loc='upper right')
    plt.show()
```
